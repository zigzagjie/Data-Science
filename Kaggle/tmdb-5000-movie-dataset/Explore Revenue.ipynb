```python
movie.drop(['homepage','tagline'],axis=1,inplace=True).
movie.dropna(inplace=True)
```

```python
movie.info()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(movie.vote_average,movie.revenue)
```

```python
import json
movie['genres']=movie.genres.apply(lambda x: json.loads(x))
movie['production_companies']=movie.production_companies.apply(lambda x: json.loads(x))
```

```python
def convertList(inputList):
    ge=[]
    for dic in inputList:
        ge.append(dic['name'])
    return ge

def getFirst(inputList):
    if len(inputList)==0:
        return np.NaN
    else:
        return inputList[0]
```

```python
g=movie.genres.apply(lambda x: getFirst(convertList(x)))
```

```python
plt.figure(figsize=(20,10))

movie.groupby(g)['revenue'].mean().plot('bar')
```

```python
plt.figure(figsize=(10,20))
sns.boxplot(movie.revenue,g)
```

```python
movie['release_date'] = pd.to_datetime(movie['release_date'])

#movie.release_date.apply(lambda x:x.year)
```

```python
year=movie.release_date.apply(lambda x:x.year)

plt.figure(figsize=(20,10))
movie.groupby(year)['revenue'].mean().plot()
```
